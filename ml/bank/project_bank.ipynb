# Отток клиентов

Из «Бета-Банка» стали уходить клиенты. Каждый месяц. Немного, но заметно. Банковские маркетологи посчитали: сохранять текущих клиентов дешевле, чем привлекать новых.

Нужно спрогнозировать, уйдёт клиент из банка в ближайшее время или нет. Вам предоставлены исторические данные о поведении клиентов и расторжении договоров с банком. 

Постройте модель с предельно большим значением *F1*-меры. Чтобы сдать проект успешно, нужно довести метрику до 0.59. Проверьте *F1*-меру на тестовой выборке самостоятельно.

Дополнительно измеряйте *AUC-ROC*, сравнивайте её значение с *F1*-мерой.

Источник данных: [https://www.kaggle.com/barelydedicated/bank-customer-churn-modeling](https://www.kaggle.com/barelydedicated/bank-customer-churn-modeling)

## Исследование задачи

```python
# Словарь для записи лучших моделей и их f1-мер
best_models = {}
```

### Разделение на выборки

```python
features = df.drop(['exited'], axis=1)
target = df['exited']

display(features.head())
display(target.head())
```

### Порядковое кодирование

```python
data_ohe = pd.get_dummies(df, drop_first=True)
features = data_ohe.drop(['exited'], axis=1)
target = data_ohe['exited']
```

<div class="alert alert-info" style="border-color: green; border-radius: 5px">
Четкой линейной зависимости каких-либо признаков не нвблюдаем, поэтому все признаки оставляем.
</div>

### Разделение на выборки

```python
features_train, features_untrain, target_train, target_untrain = train_test_split(features, target, test_size=0.4, random_state=12345)
display(features_train)
display(features_untrain.head())
display(target_train.head())
display(target_untrain.head())
```

```python
features_valid, features_test, target_valid, target_test = train_test_split(features_untrain, target_untrain, test_size=0.5, random_state=12345)
display(features_valid.head())
display(features_test.head())
display(target_valid.head())
display(target_test.head())
```

### Масштабирование

```python
df.info()
```

```python
numeric = ['credit_score', 'age', 'balance', 'estimated_salary']

scaler = StandardScaler()
pd.options.mode.chained_assignment = None
scaler.fit(features_train[numeric])
features_train[numeric] = scaler.transform(features_train[numeric])
features_valid[numeric] = scaler.transform(features_valid[numeric])

display(features_train.head())
display(features_valid.head())
```

### Обучение моделей без учета дисбаланса

#### Дерево решений

```python
best_model = None
best_f1 = 0
best_depth = 0

for depth in range(1, 50):
    model_tree = DecisionTreeClassifier(random_state=12345, max_depth=depth)
    model_tree.fit(features_train, target_train)
    predictions_valid = model_tree.predict(features_valid)
    f1 = f1_score(target_valid, predictions_valid)
    if f1 > best_f1:
        best_f1 = f1
        best_depth = depth
        best_model = model_tree

best_models[best_model] = best_f1
print('f1', best_f1)
print('best_depth_tree', best_depth)
```

#### Логистическая регрессия

```python
model_linear = LogisticRegression(random_state=12345, solver='lbfgs', max_iter=100)
model_linear.fit(features_train, target_train)
predictions_valid = model_linear.predict(features_valid)
f1 = f1_score(target_valid, predictions_valid)

best_models[model_linear] = f1
print('f1', f1)
```

#### Случайный лес

```python
def search_forest_depth(features, target):
    best_f1_forest = 0
    best_depth_forest = 0

    for depth in range(1, 200):
        model_forest = RandomForestClassifier(max_depth=depth, n_estimators=20, random_state=12345)
        model_forest.fit(features, target)
        predictions_valid = model_forest.predict(features_valid)
        f1 = f1_score(target_valid, predictions_valid)
        if f1 > best_f1_forest:
            best_f1_forest = f1
            best_depth_forest = depth

    return best_depth_forest
```

```python
best_depth_forest = search_forest_depth(features_train, target_train)
print('best_depth_forest', best_depth_forest)
```

```python
best_model = None
best_f1 = 0
best_est = 0

for est in range(1, 100):
    model_forest = RandomForestClassifier(max_depth=best_depth_forest , n_estimators=est, random_state=12345)
    model_forest.fit(features_train, target_train)
    predictions_valid = model_forest.predict(features_valid)
    f1 = f1_score(target_valid, predictions_valid)
    if f1 > best_f1:
        best_f1 = f1
        best_est = est
        best_model = model_forest

best_models[best_model] = best_f1
print('f1', f1)
print('best_est', best_est)
```

```python
# Также подберем параметры с помощью GridSearchCV
def grid_model(features, target, class_weight = None):
    model_forest_grid = RandomForestClassifier()
    params = {
        'max_depth': [randint(1, 100)],
        'n_estimators': [randint(1, 100)],
        'random_state': [random_state],
        'class_weight': [class_weight]
    }

    grid = GridSearchCV(model_forest_grid, params)
    grid.fit(features, target)
    print(grid.best_params_)
```

```python
grid_model(features_train, target_train)
```

```python
model_forest = RandomForestClassifier(max_depth=50, n_estimators=43, random_state=random_state)
model_forest.fit(features_train, target_train)
predictions_valid = model_forest.predict(features_valid)
f1 = f1_score(target_valid, predictions_valid)

best_models[model_forest] = f1
print('f1', f1)
```

### Вывод

Наилучшей моделью без учета баланса классов оказазась модель `Случайный лес` с максимальной глубиной 50 и 43 деревьями.

### Увеличение выборки

```python
def upsample(features, target, repeat):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_upsampled = pd.concat([features_zeros] + [features_ones] * repeat)
    target_upsampled = pd.concat([target_zeros] + [target_ones] * repeat)
    
    features_upsampled, target_upsampled = shuffle(
        features_upsampled, target_upsampled, random_state=12345)
    
    return features_upsampled, target_upsampled
```

```python
features_train_upsampled, target_train_upsampled = upsample(features_train, target_train, 3)
display(features_train_upsampled.head())
display(target_train_upsampled.head())
```

```python
best_model = None
best_f1 = 0
best_est = 0

for est in range(1, 100):
    model_forest = RandomForestClassifier(max_depth=best_depth_forest , n_estimators=est, random_state=12345)
    model_forest.fit(features_train_upsampled, target_train_upsampled)
    predictions_valid = model_forest.predict(features_valid)
    f1 = f1_score(target_valid, predictions_valid)
    if f1 > best_f1:
        best_f1 = f1
        best_est = est
        best_model = model_forest

best_models[best_model] = best_f1
print('f1', f1)
print('best_est', best_est)
```

```python
def show_best_model():    
    for key, value in best_models.items():
        if value == max(best_models.values()):
            print(f'Лучшая модель - {key} с F1-мерой, равной {value}')
            return key
```

```python
best_model = show_best_model()
```
