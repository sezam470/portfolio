# Сбор и подготовка данных для проверки гипотез



Необходимо организовать одинаковый сбор данных (в одинаковых условиях, единицах измерения, для номинальных данных - выбором из списка, а не в свободной форме) и возможность их аггрегации.

**Типы данных:**
- первичные (собранные впервые для решения текущей задачи)
- вторичные (собранные для других целей)

**По наличию структуры:**
- неструктурированные (тексты на естественном языке)
- слабоструктурированные (сообщение электронной почты)
- структурированные (списки, множества, массивы, ряды, таблицы, транзакции)

**Виды данных:**
- фактоиды
- ряды
- таблицы
- транзакции (самый полезный вид данных)

**Источники данных:**
- результаты прямых измерений
(это лучшие данные, так как они наименее подвержены субъективным искажениям, это сырые транзакции без всякой агрегации, собранные датчиками, но они требуют предаварительной обработки перед анализом - согласование идентификации объектов в разных системах, очистки от шумов, восстановление пропущенных значений и тп)
- результаты социологических опросов
(но необходимо наличие правильно составленных анкет для опроса, чтобы ответы были всеобъемлющими и взаимоисключающими; надежного инструментария для заполенения и анализа анкет; создание благоприятной, психологически комфортной обстановки для заполнения и анализа анкет)
- документы (как правило слабо структурированы -  имеют свободную форму, - или неразборчивые)
- социальные сети (хороши для анализа социальной активности и ее прогнозирования)
- внешние источники (поисковые системы, хранилища данных, правительственные базы данных, исследовательские учреждения; необходимо ссылаться на источник - автора/владельца + название + ссылка)
- наблюдения (особенно если объекты наблюдения - не люди, и если максимально исключен человеческий фактор)

**Задачи:**
1. Сбор всей генеральной совокупности (если возможно) или формирование репрезентативной выборки
2. Заполнение пропущенных значений
3. Проверка правдоподобности данных
4. Объединение данных из разных источников:
- провести соответствие полей (*какие поля одной таблицы соответствуют полям другой таблицы)
- переименование полей к единому стилю
- обработка полей, которые встречаются не во всех источниках (удалить или наоборот расширить данные, а на этапе анализа заполнить значениями-заглушками или средними значениями)
- исключить значения вне диапазонов
5. Вычищение из исходных файлов всех нечитаемых символов
5. Загрузка данных в хранилище
(если данных очень много, то загрузку производить непосредственно на сервере, где расположено хранилище, и загружать все исходные данные как текстовые поля)

# Очистка данных от шумов и аномалий

Избавление данных от шумов и аномалий:
1. Устранение дубликатов
2. Проверка шаблонов (сравнение с регулярными выражениями):
- попытаться получить адекватные данные
- заполнить значения явно заданными символами, означающими неопределенное значение
- или вообще исключить из анализа

Про регулярные выражения: https://support.google.com/analytics/topic/1034375?hl=ru&ref_topic=1032939
3. Контроль диапазонов:
- задаются предметной областью (лучше всего)


# Преобразование данных

1. Конвертация данных к единому формату (таблица, JSON, XML)
2. Приведение чисел в строковом формате к численному формату.
3. Разделение данных (*разделение имени и фамилии путем сравнения со справочником нестандартных имен и др.)
4. Приведение к одним шкалам с одинаковыми единицами измерения (*тонны в кг, минуты в секунды и др.)
5. Нормировка переменных - приведение к диапазону (*от 0 до 1 и др.)
6. Приведение к унифицированной лексике (относится к номинальным данным, если при формировании данных значения не выбирались из списка, а заполнялись свободно):
- перевести все значения к единому регистру
- выявить список различающихся значений, на основе которого составить список соответствий
- произвести замену значений
- выявить скрытые дубликаты (по уникальным признакам объекта - *паспорту, номеру телефона, электронному адресу), так как в разных источниках может быть разная система идентификации объектов
7. Заполнение отсутствующих числовых значений:
- выделение таких полей специальными значениями и исключение из дальнейшего анализа
- аппроксимация пропущенных значений на основе существующих данных (*а) средним значением всего набора данных или б) в случае временных рядов - методом скользящего окна из ближайших соседей или в) на основе исторических значений - соотношений временных интервалов и др.)
